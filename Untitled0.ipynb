```python
import numpy as np

np.random.seed(0)
n_estudiantes = 6500

# Código: 2 dígitos de carrera + 2 dígitos de año + 3 dígitos de ID
codigos = np.array([
    f"{np.random.randint(1, 100):02d}{np.random.randint(80, 100):02d}{np.random.randint(0,100):03d}"
    for _ in range(n_estudiantes)
])

nombres = np.array([f"Estudiante_{i}_" for i in range(n_estudiantes)])
promedios = np.round(np.random.uniform(2.0, 5.0, n_estudiantes), 2)

codigos_carrera = np.array([c[:2] for c in codigos])
carreras_unicas = np.unique(codigos_carrera)

print("Carreras existentes y estudiantes con promedio mayor o igual a 4:\n")

for carrera in carreras_unicas:
    filtro = (codigos_carrera == carrera) & (promedios >= 4.0)
    indices = np.where(filtro)[0]

    if len(indices) > 0:
        print(f"Carrera {carrera}: {len(indices)} estudiantes con promedio >= 4")
        for i in indices:
            print(f"  Código: {codigos[i]}, Nombre: {nombres[i]}, Promedio: {promedios[i]}")
        print()

años_2d = np.array([int(c[2:4]) for c in codigos])
años_ingreso = np.where(años_2d >= 80, 1900 + años_2d, 2000 + años_2d)


filtro_condicionales = (años_ingreso < 1990) & (promedios >= 2.7) & (promedios < 3.2)
indices_condicionales = np.where(filtro_condicionales)[0]

print("Estudiantes ingresados antes de 1990 y condicionales (promedio entre 2.7 y 3.2):\n")
for i in indices_condicionales:
    print(f"Código: {codigos[i]}, Nombre: {nombres[i]}, Año ingreso: {años_ingreso[i]}, Promedio: {promedios[i]}")

```
